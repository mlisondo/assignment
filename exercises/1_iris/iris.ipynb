# Iris Dataset Analysis

## Prepare the dataset

Use `numpy` or to load `data/iris.data`. Remember to look at the file contents to understand the data format.

```python
import numpy as np
import matplotlib.pyplot as plt

attributes = np.loadtxt("data/iris.attributes", delimiter=",")
classes = np.loadtxt('data/iris.classes', dtype=str)
```

Create a map that returns the attribute name given the attribute index.

- 0 -> sepal length
- 1 -> sepal width
- 2 -> petal length
- 3 -> petal width

```python
map = {
    0 : 'sepal length',
    1 : 'sepal width',
    2 : 'petal length',
    3 : 'petal width'
}
```

## Data analysis

Write a function that prints the mean and another that prints the standard deviation of each attribute.

```python
def mean(data):
    print("".center(40, '~'))
    print("Mean of".center(40, ' '))
    print("".center(40, '~'))
    result = []
    for i in range(4):
        mean_val = np.mean(data[:, i])
        print(f"{map[i]} = {mean_val}")
        result.append(mean_val)
    print()
    return np.array(result)

def std(data):
    print("".center(40, '~'))
    print("Standard deviation of".center(40,' '))
    print("".center(40, '~'))
    result = []
    for i in range(4):
        std_val = np.std(attributes[:,i])
        print(f"{map[i]} = {std_val}")
        result.append(std_val)
    print()
    return np.array(result)
```

Use the functions you defined above to print the mean of the attributes for each class.

```python
mean(attributes)
```

```python
std(attributes)
```

Now, let's plot this data. Use `matplotlib` to create scatter plots for petal and sepal. Use different colors for each class.

```python
setosa = classes == 'Iris-setosa'
versicolor = classes == 'Iris-versicolor'
virginica = classes == 'Iris-virginica'

set_s_length = attributes[setosa, 0]
set_s_width = attributes[setosa, 1]
set_p_length = attributes[setosa, 2]
set_p_width = attributes[setosa, 3]

ver_s_length = attributes[versicolor, 0]
ver_s_width = attributes[versicolor, 1]
ver_p_length = attributes[versicolor, 2]
ver_p_width = attributes[versicolor, 3]

vir_s_length = attributes[virginica, 0]
vir_s_width = attributes[virginica, 1]
vir_p_length = attributes[virginica, 2]
vir_p_width = attributes[virginica, 3]
```

```python
plt.scatter(set_p_length, set_p_width, label="Setosa", color="red")
plt.scatter(ver_p_length, ver_p_width, label="Versicolor", color="green")
plt.scatter(vir_p_length, vir_p_width, label="Virginica", color="blue")

plt.xlabel("Petal Length")
plt.ylabel("Petal Width")
plt.legend()
plt.title("Iris Petal Distribution")
plt.show()
```
